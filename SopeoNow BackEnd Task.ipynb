#### Printing DataFrame after removing duplicate rows and removing row which has all empty columns (ie., NaN values).

```python
df = pd.read_csv(r'C:\Users\kalee\python\Assignement\dataset.csv')
df = df.drop_duplicates()
df.dropna(axis = 0, how = 'all', inplace = True)
with pd.option_context('display.max_rows', None, 'display.max_columns', None): 
    display(df)
```

```python
d = df.groupby('NURSEKEY').apply(lambda x: x['NURSENAME'].unique()) #Each Nurse key with respect to Nurse name
dd= df.groupby('ROOMKEY').apply(lambda x: x['ROOMNAME'].unique())   #Each Room key with respect to Room name
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(d)
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(dd)
```

# Case 1: ID's ('BUILDINGKEY' / 'FLOORKEY' / 'NURSEKEY' / 'BEDKEY') are Unique and irrespective of building, floor, ward, bed respectively (considering as independent to each other).

## 1.Get the Count of beds in each ward.

```python
df1 = df.groupby('NURSEKEY').apply(lambda x: x['BEDKEY'].nunique())
with pd.option_context('display.max_rows', None, 'display.max_columns', None): 
    display(df1)
```

## 2.Get the count of bed in each room.

```python
df2 = df.groupby('ROOMKEY').apply(lambda x: x['BEDKEY'].nunique())
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  
    display(df2)
```

## 3.Get the count of the bed under each status.

```python
df3 = df.groupby('BEDSTATUS').apply(lambda x: x['BEDKEY'].count()).reset_index(name='Beds')
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  
    display(df3)
```

#### There are duplicate Bedkeys in each Bedstatus
#### So we take unique Bedkeys in each Bedstatus by using nunique function instead of count function

```python
df3 = df.groupby('BEDSTATUS').apply(lambda x: x['BEDKEY'].nunique()).reset_index(name='Beds')
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  
    display(df3)
```

## 4.Calculate the overall occupancy percentage (total occupied / total bed).

```python
total_beds = df3['Beds'].sum()
total_occupied = df3.loc[df3['BEDSTATUS'].isin(['Occupied Discharge', 'Occupied(Female)', 'Occupied(Male)']), 'Beds'].sum()
print("Total Beds", total_beds)
print("Total Occupied", total_occupied)
print("Occupancy Percentage" , (total_occupied * 100)/total_beds)
```

## 7.Calculate count of bed with all the different bed statuses for each ward.

```python
d1 = df.groupby(['NURSEKEY', 'BEDSTATUS']).apply(lambda x: x['BEDKEY'].nunique())
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(d1)
```

## 5.Calculate occupancy for each ward.

```python
d11 = d1.reset_index(name='Beds')
totalUnderWard = d11.groupby('NURSEKEY')['Beds'].sum().reset_index(name='Total Beds')

occupiedTotal =   (d11.assign(occupiedcount = d11['Beds'].where(d11['BEDSTATUS'].isin(['Occupied Discharge', 'Occupied(Female)', 'Occupied(Male)'])))
    .groupby(['NURSEKEY'])
    .agg(
        Occupied_Total=('occupiedcount', 'sum')
        
    )).reset_index()

totalUnderWard['Occupied Total'] = occupiedTotal['Occupied_Total'].astype(int)
totalUnderWard ["Occupancy Percentage"] = totalUnderWard ["Occupied Total"]*100 /totalUnderWard ["Total Beds"]
print(totalUnderWard)
```

## 8.Calculate count of bed with all the different bed statuses for each room.

```python
d2 = df.groupby(['ROOMKEY', 'BEDSTATUS']).apply(lambda x: x['BEDKEY'].nunique())
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(d2)
```

## 6.Calculate occupancy for each room.

```python
d22 = d2.reset_index(name='Beds')
totalUnderRoom = d22.groupby('ROOMKEY')['Beds'].sum().reset_index(name='Total Beds')

occupiedTotal =   (d22.assign(occupiedcount = d22['Beds'].where(d22['BEDSTATUS'].isin(['Occupied Discharge', 'Occupied(Female)', 'Occupied(Male)'])))
    .groupby(['ROOMKEY'])
    .agg(
        Occupied_Total=('occupiedcount', 'sum')
        
    )).reset_index()

totalUnderRoom['Occupied Total'] = occupiedTotal['Occupied_Total'].astype(int)
totalUnderRoom ["Occupancy Percentage"] = totalUnderRoom ["Occupied Total"]*100 /totalUnderRoom ["Total Beds"]
print(totalUnderRoom)
```

##### Bed is dependent on Building, Floor, Ward and Room.
##### Room is dependent on Building, Floor and Ward.
##### Ward is dependent on Building and Floor.
